```python
import pandas as pd
import numpy as np

# Load datasets
train_data = pd.read_csv('Datasets/train.csv')
test_data = pd.read_csv('Datasets/test.csv')

# Checking for missing values
print(train_data.isnull().sum())
print(test_data.isnull().sum())

# 1. Handling missing 'Age' with median
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)

# 2. Handling missing 'Cabin' (drop or create a flag for missing cabins)
train_data['Cabin'] = train_data['Cabin'].fillna('Unknown')
test_data['Cabin'] = test_data['Cabin'].fillna('Unknown')

# 3. Handling missing 'Embarked' with the mode (most frequent value)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)

# 4. Filling 'Fare' missing in test data with median
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)

# Confirm no missing data remains
print(train_data.isnull().sum())
print(test_data.isnull().sum())
```

# Feature Engineering

```python
# 1. Feature: Family Size
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1
test_data['FamilySize'] = test_data['SibSp'] + test_data['Parch'] + 1

# 2. Feature: IsAlone (1 if the passenger is alone, 0 otherwise)
train_data['IsAlone'] = np.where(train_data['FamilySize'] == 1, 1, 0)
test_data['IsAlone'] = np.where(test_data['FamilySize'] == 1, 1, 0)

# 3. Extracting titles from names
train_data['Title'] = train_data['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())
test_data['Title'] = test_data['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())

# Simplify rare titles
title_mapping = {
    "Mr": "Mr", "Miss": "Miss", "Mrs": "Mrs", "Master": "Master",
    "Dr": "Rare", "Rev": "Rare", "Col": "Rare", "Major": "Rare",
    "Mlle": "Miss", "Ms": "Miss", "Mme": "Mrs", "Don": "Rare",
    "Lady": "Rare", "Countess": "Rare", "Jonkheer": "Rare",
    "Sir": "Rare", "Capt": "Rare", "Dona": "Rare"
}
train_data['Title'] = train_data['Title'].map(title_mapping)
test_data['Title'] = test_data['Title'].map(title_mapping)

# 4. Convert categorical features like 'Sex', 'Embarked', and 'Title' into numerical
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})

train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
test_data['Embarked'] = test_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Final check of data
print(train_data.head())
print(test_data.head())
```

# Exploratory Data Analysis (EDA)

```python
# 3. Survival rate by Family Size
sns.barplot(x='FamilySize', y='Survived', data=train_data)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
# 5. Survival rate by Title
sns.barplot(x='Title', y='Survived', data=train_data)
plt.title('Survival Rate by Title')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Age distribution of survivors
plt.figure(figsize=(10, 6))
sns.kdeplot(train_data[train_data['Survived'] == 1]['Age'], label='Survived', shade=True, color="g")
sns.kdeplot(train_data[train_data['Survived'] == 0]['Age'], label='Not Survived', shade=True, color="r")
plt.title('Age Distribution - Survived vs Not Survived')
plt.legend()
plt.show()
```
